```python
# Filter columns
df_2=df_1.filter(items=['Studyplatform'])
print(df_2)
```

```python
#Converting csv Dataframe to String
Df_to_st= df_2.to_string()[1:].replace('  ', '') + '\n'
Df_to_st
```

```python
#Removing number, punctuation, and extra spaces
import string
import re
def clean_text(text):
    # remove numbers
    text_nonum = re.sub(r'\d+', '', Df_to_st)
    # remove punctuations and convert characters to lower case
    text_nopunct = "".join([char.lower() for char in text_nonum if char not in string.punctuation])
    # substitute multiple whitespace with single whitespace
    # Also, removes leading and trailing whitespaces
    text_no_doublespace = re.sub('\s+', ' ', text_nopunct).strip()
    return text_no_doublespace


cleaned_text = clean_text(Df_to_st)
cleaned_text
```

```python
#Removing Stopwords
import nltk
from nltk.corpus import stopwords

from nltk.tokenize import word_tokenize

#Removing stopwordws
stop_words = set(stopwords.words('english'))
word_tokens = word_tokenize(cleaned_text)
  
filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]
  
filtered_sentence = []
  
for w in word_tokens:
    if w not in stop_words:
        filtered_sentence.append(w)
print(filtered_sentence)
```

```python
#Converting the LIST of words that does not contain stopwords anymore into STRING
string = " ".join(filtered_sentence)
string
```

```python
#Frequency of words in the string
string_1 = string.split()

wordfreq = []
for w in string_1:
    wordfreq.append(string_1.count(w))

print("Pairs\n" + str(list(zip(string_1, wordfreq))))
```

```python
# To find the k most frequent words
# from data set
from collections import Counter

# split() returns list of all the words in the string
split_it = string.split()
  
# Pass the split_it list to instance of Counter class.
Counter = Counter(split_it)
  
# most_common() produces k frequently encountered
# input values and their respective counts.
most_occur = Counter.most_common(20)
print(most_occur)
```
